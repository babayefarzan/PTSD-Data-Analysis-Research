```python
import numpy as np
import pandas as pd
%matplotlib inline
import glob
import keras
from functools import reduce
```

```python
path = r'/Users/mahnooshsadeghi/Desktop/PTSD/PTSD Data/Windowed Data' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
li=[]
frame=[]
```

```python
for filename in all_files:
    df1 = pd.read_csv(filename, index_col=None, header=0)
    linaccx=df1.groupby('windowno', as_index=False)['linear_accel_x'].mean()
    linaccy=df1.groupby('windowno', as_index=False)['linear_accel_y'].mean()
    linaccz=df1.groupby('windowno', as_index=False)['linear_accel_z'].mean()
    df1 ['linacc']= (df1['linear_accel_x']**2+df1['linear_accel_y']**2+df1['linear_accel_z']**2)**0.5
    linaccmin=df1.groupby('windowno', as_index=False)['linacc'].min()
    linaccmin=linaccmin.rename(columns={"linacc": "linaccmin"})
    linaccmax=df1.groupby('windowno', as_index=False)['linacc'].max()
    linaccmax=linaccmax.rename(columns={"linacc": "linaccmax"})
    linaccsd=df1.groupby('windowno').agg(np.std,ddof=0)['linacc']
    linaccsd=pd.DataFrame(linaccsd)
    linaccsd=linaccsd.rename(columns={"linacc": "linaccsd"})
    linaccmean=df1.groupby('windowno', as_index=False)['linacc'].mean()
    linaccmean=linaccmean.rename(columns={"linacc": "linaccmean"})
    stress=df1.groupby('windowno', as_index=False)['ptsd_moment'].max()
    participant = df1['participant'].iloc[0]
    data=reduce(lambda x,y: pd.merge(x,y, on='windowno', how='outer'),
       [linaccmean,linaccmin,linaccmax,linaccsd,linaccx,linaccy,linaccz,stress])
    data['participant']= participant
    li.append(data)
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame['linaccrange']= frame['linaccmax']-frame['linaccmin']
```
